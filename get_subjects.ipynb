```python
import requests
from bs4 import BeautifulSoup
import re
import pandas as pd
import json
from uO_scrape import scrape_subjects, get_course_tables, get_courses, extract_credits, extract_codes
```

# Subjects

```python
subjects = scrape_subjects()
subjects.head()
```

```python
if len(subjects) != len(set(subjects.Code)):
    print("WARNING! NON-UNIQUE SUBJECT CODES HAVE BEEN DETECTED.")
```

```python
subjects.to_csv("uOttawa_subjects.csv", index=False)
```

# Course Description

```python
links = subjects.Link.tolist()

course_tables = get_course_tables(links)
```

```python
course_tables['mat']
```

```python
json_courses = dict()
for key in course_tables:
    json_courses[key] = course_tables[key].to_json(orient='records')

with open("uOttawa_courses.json", 'w') as f:
    json.dump(json_courses, f, indent='\t')
```

```python
all_courses = pd.concat([course_tables[key] for key in course_tables])
print(len(all_courses))
all_courses.to_csv('uOttawa_courses.csv', index=False)
all_courses.head()
```
